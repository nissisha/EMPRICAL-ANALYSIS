```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(14, 6))
plt.style.use('fivethirtyeight')

ax = sns.countplot(x='Product Category', data=data, palette='Set1')

ax.set_xlabel(xlabel="Product Category", fontsize=18)
ax.set_ylabel(ylabel="Count", fontsize=18)
ax.set_title(label="Most Popular Product Category", fontsize=20)

plt.xticks(rotation=45) 
plt.show()plt.figure(figsize=(14, 6))
plt.style.use('fivethirtyeight')

ax = sns.barplot(x='Product Category', y='Quantity ordered new', data=data, estimator=sum, ci=None, palette='Set2')

ax.set_xlabel(xlabel="Product Category", fontsize=18)
ax.set_ylabel(ylabel="Total Quantity Ordered", fontsize=18)
ax.set_title(label="Total Quantity Ordered by Product Category", fontsize=20)

plt.xticks(rotation=45)
plt.show()
```
